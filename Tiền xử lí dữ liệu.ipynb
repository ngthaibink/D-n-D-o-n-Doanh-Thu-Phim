```python
import pandas as pd

# 1. Đọc dữ liệu
df = pd.read_csv("movies.csv")

# 2. Kiểm tra kích thước dữ liệu
print("Kích thước dữ liệu:", df.shape)

# 3. Kiểm tra kiểu dữ liệu
print("\nThông tin cột:")
print(df.info())

# 4. Kiểm tra missing values
print("\nMissing values:")
print(df.isnull().sum())

# 5. Kiểm tra dữ liệu rác (budget = 0, gross = 0)
print("\nSố dòng budget = 0:", (df["budget"] == 0).sum())
print("Số dòng gross = 0:", (df["gross"] == 0).sum())

# 6. Loại bỏ dòng không hợp lệ
df = df[(df["budget"] != 0) & (df["gross"] != 0)]

# 7. Chuẩn hóa dữ liệu
df["genre"] = df["genre"].str.strip()
df["rating"] = df["rating"].str.upper()

# 8. Đảm bảo year là kiểu số
df["year"] = pd.to_numeric(df["year"], errors="coerce")

# 9. Tính Profit
df["Profit"] = df["gross"] - df["budget"]

# 10. Xuất file clean_movies.csv
df.to_csv("clean_movies.csv", index=False)

print("\nHoàn thành làm sạch dữ liệu!")
```
